```python
import pandas as pd 

smoothing_1e10_path = "/shares/grossniklaus.botinst.uzh/eharputluoglu/calculation_output/methylation_analysis_results_3.csv"
smoothing_1_path = "/shares/grossniklaus.botinst.uzh/eharputluoglu/calculation_output/methylation_analysis_results_Smooting_1.csv" 
smoothing_1_10_path = "/shares/grossniklaus.botinst.uzh/eharputluoglu/calculation_output/methylation_analysis_results_Smooting_1e-10.csv" 
smoothing_1_5_path = "/shares/grossniklaus.botinst.uzh/eharputluoglu/calculation_output/methylation_analysis_results_Smooting_1e-5.csv" 



smoothing_1e10 = pd.read_csv(smoothing_1e10_path)
smoothing_1 = pd.read_csv(smoothing_1_path)
smoothing_1e_10 = pd.read_csv(smoothing_1e10_path)
smoothing_1e_5 = pd.read_csv(smoothing_1_5_path)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

min_val = min(smoothing_1e10["SGJS Divergence"].min(), smoothing_1["SGJS Divergence"].min())
max_val = max(smoothing_1e10["SGJS Divergence"].max(), smoothing_1["SGJS Divergence"].max())

plt.figure(figsize=(8, 6)) 
plt.scatter(smoothing_1e10["SGJS Divergence"], smoothing_1["SGJS Divergence"], alpha=0.7) 
plt.xlim([min_val, max_val])
plt.ylim([min_val, max_val])

# Perfect Correlation Line
x = np.linspace(min_val, max_val, 100)
y = x
plt.plot(x, y, color='red', linestyle='dashed', linewidth=1, label='Perfect Correlation (y=x)')

plt.xlabel("Smoothing 1e10 - GJS Divergence", fontsize=12)
plt.ylabel("Smoothing 1 - GJS Divergence", fontsize=12)
plt.title("Scatter Plot of GJS Divergence", fontsize=14)
plt.legend(fontsize=10)
plt.grid(alpha=0.4) 

plt.show()

```
